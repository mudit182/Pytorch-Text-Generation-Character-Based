```python
import os
import string

from PyPDF2 import PdfReader
```

```python
pdfs_to_extract = [file for file in os.listdir('pdfs') if file.endswith('.pdf')]

print(f'Extracting text from {len(pdfs_to_extract)} files')
```

```python
# creating a pdf reader object
# reader = PdfReader('pdfs/01EarlyCulturalWritings.pdf')

text = ''
for file_name in pdfs_to_extract:
    try:
        reader = PdfReader(f'pdfs/{file_name}')
        print(f'{file_name}: {len(reader.pages)} pages')
        for page in reader.pages:
            text += page.extract_text()
        print(f'{file_name}: Finished text extraction!')
    except Exception as e:
        print(f'ERROR --- {file_name}!')
        print(e)
```

```python
# Clean text - remove unknown characters
from cleantext import clean

cleaned_text = clean(text,
    fix_unicode=True,               # fix various unicode errors
    to_ascii=True,                  # transliterate to closest ASCII representation
    lower=False,                     # lowercase text
    no_line_breaks=False,           # fully strip line breaks as opposed to only normalizing them
    no_urls=False,                  # replace all URLs with a special token
    no_emails=False,                # replace all email addresses with a special token
    no_phone_numbers=False,         # replace all phone numbers with a special token
    no_numbers=False,               # replace all numbers with a special token
    no_digits=False,                # replace all digits with a special token
    no_currency_symbols=False,      # replace all currency symbols with a special token
    no_punct=False,                 # remove punctuations
    lang="en"                       # set to 'de' for German special handling
)
```

```python
with open('sri_aurobindo_all_lines.txt', 'w') as file:
    file.write(cleaned_text)
```

```python
accepted_chars = string.ascii_letters + string.digits + string.punctuation + string.whitespace

final_text = ''
for char in text:
    if char in accepted_chars:
        final_text += char

print(len(text), len(final_text))
```

```python
with open('sri_aurobindo_all_lines_final.txt', 'w') as file:
    file.write(final_text)
```
